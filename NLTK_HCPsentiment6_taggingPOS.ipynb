```python
"""
KEEP THIS (part1)
"""

class rawText(object):
    def __init__(self, name_file):
        self.name_file = name_file
        self.lines = None
        self.idx = 0

    def read_file(self):
        """Read the file content and store inside class instance"""
        with open(self.name_file, 'r') as in_file:
            self.lines = [x.rstrip().replace('“','').replace('”',"").replace('"','')
                          .replace('[','(').replace(']',')').replace("’",'')
                          .replace("—",'-').replace("…",' ').replace("£",'GBP').replace("–",'-') for x in in_file.readlines()]
        return self.lines

    def __next__(self):
        if not self.lines:
            self.read_file()
        try: 
            self.idx += 1
            return self.lines[self.idx - 1].rstrip().replace('“','').replace('”',"").replace('"','').replace('[','(').replace(']',')').replace("’",'').replace("—",'-').replace("…",' ').replace("£",'GBP').replace("–",'-')
        except IndexError:
            raise StopIteration

    def __iter__(self): 
        return self

    # replaces your display_file
    def __str__(self):
        return self.name_file + (" : " if self.lines else "") + (
                                 "    ".join(self.lines or []))
```
